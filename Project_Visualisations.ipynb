# Multi-Agent Deep Reinforcement Learning for Liquidation Strategy Analysis - Visualisations

```python
import utils
import numpy as np
import pandas as pd
import seaborn as sns
```

## Theorem 1

```python
m1_shortfall_list = np.load('Project_Results/1e-06_shortfall_list_10000.npy')
m03_shortfall_list = np.load('Project_Results/1e-06_shortfall_list_0.3M_10000.npy')
m07_shortfall_list = np.load('Project_Results/1e-06_shortfall_list_0.7M_10000.npy')

E1_list = []
E2_list = []
B1_list = []
B2_list = []
for i in range(len(m1_shortfall_list)):
    E1_list.append(m1_shortfall_list[i])
    E2_list.append(m03_shortfall_list[i] + m07_shortfall_list[i])
    B1_list.append(m03_shortfall_list[i])
    B2_list.append(m07_shortfall_list[i])

df1 = pd.DataFrame()
df1['Agent A'] = E1_list
df1['B1 + B2'] = E2_list
df1['Agent B1'] = B1_list
df1['Agent B2'] = B2_list

sns.set()
ax = sns.lineplot(data = df1)
ax.set_ylabel('Expected Shortfall',fontsize = 20)
ax.set_xlabel('Episodes ',fontsize = 20)
ax.set_xticklabels([0,0,2000,4000,6000,8000,10000],fontsize = 14)
ax.grid(visible=True, which='major', color='black', linewidth=0.5,linestyle = '--')
```

## Theorem 2

```python
A1 = np.load('Project_Results/1e-04_optimal_trajectory_10000.npy')
A2 = np.load('Project_Results/1e-09_optimal_trajectory_10000.npy')
B = np.load('Project_Results/1e-04_1e-09_trajectory_10000.npy')
A1_list = []
A2_list = []
B1_list = []
B2_list = []
for i in range(len(A1)):
    A1_list.append(A1[i])
    A2_list.append(A2[i])
    B1_list.append(B[i][0])
    B2_list.append(B[i][1])
    
df2 = pd.DataFrame()
df2['Agent A1'] = A1_list
df2['Agent A2'] = A2_list
df2['Agent B1'] = B1_list
df2['Agent B2'] = B2_list

ax = sns.lineplot(data = df2)
ax.set_ylabel('Percentage',fontsize = 20)
ax.set_xlabel('Time (Day)',fontsize = 20)
ax.set_xticklabels([0,0,10,20,30,40,50,60], fontsize = 14)
ax.grid(visible=True, which='major', color='black', linewidth=0.5,linestyle = '--')
```

## Cooperation vs Competition

```python
A = np.load('Project_Results/1e-06_shortfall_list_10000.npy')
B = np.load('Project_Results/1e-06_1e-06_competition_shortfall_list_10000.npy')
C = np.load('Project_Results/1e-06_1e-06_cooperation_shortfall_list_10000.npy')
A_list = []
B_list = []
C_list = []
for i in range(len(A)):
    A_list.append(A[i])
    B_list.append(B[i].sum())
    C_list.append(C[i].sum())

df3 = pd.DataFrame()
df3['Independent'] = A_list
df3['Competitive'] = B_list
df3['Cooperative'] = C_list

ax = sns.lineplot(data = df3)
ax.set_ylabel('Expected Shortfall',fontsize = 20)
ax.set_xlabel('Episodes',fontsize = 20)
ax.set_xticklabels([0,0,2000,4000,6000,8000,10000],fontsize = 14)
ax.grid(visible=True, which='major', color='black', linewidth=0.5,linestyle = '--')
```
